```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt
import numpy as np

# Simular datos (20 registros)
np.random.seed(42)

horas_estudiadas = np.random.uniform(1, 10, 20)
puntuacion_examen = horas_estudiadas * np.random.uniform(7, 10) + np.random.normal(0, 5, 20)
nivel_motivacion = np.random.randint(10, 100, 20)

# Clasificar índices por nivel de motivación
baja_idx = nivel_motivacion <= 33
media_idx = (nivel_motivacion > 33) & (nivel_motivacion <= 66)
alta_idx = nivel_motivacion > 66

# Crear gráfico de dispersión por grupo
plt.figure(figsize=(10, 6))

plt.scatter(horas_estudiadas[baja_idx], puntuacion_examen[baja_idx], color='red', label='Baja motivación', marker='o', s=80)
plt.scatter(horas_estudiadas[media_idx], puntuacion_examen[media_idx], color='orange', label='Media motivación', marker='s', s=80)
plt.scatter(horas_estudiadas[alta_idx], puntuacion_examen[alta_idx], color='green', label='Alta motivación', marker='^', s=80)

# Etiquetas y título
plt.xlabel('Horas estudiadas')
plt.ylabel('Puntuación en el examen')
plt.title('Relación entre horas de estudio, puntuación y motivación')
plt.legend()

# Mostrar gráfico
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear el DataFrame con los datos de ventas mensuales por tipo de vehículo

datos = pd.DataFrame({
'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
'SUV': [120, 130, 135, 140, 150, 160, 170, 180, 175, 185, 190, 200],
'Sedan': [100, 90, 95, 110, 120, 130, 140, 145, 150, 160, 170, 180],
'Hatchback': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125, 130, 135]
})

# Tamaño
plt.figure(figsize=(12, 6))

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Crear el gráfico de líneas con Matplotlib
plt.plot(datos['Mes'], datos['SUV'], label='SUV', marker='o', color='blue', linestyle='--')
plt.plot(datos['Mes'], datos['Sedan'], label='Sedan', marker='s', color='green', linestyle='-.')
plt.plot(datos['Mes'], datos['Hatchback'], label='Hatchback', marker='^', color='red', linestyle=':')

# Personalizar el gráfico con títulos y etiquetas

plt.title('Ventas Mensuales de Vehículos por Tipo', fontsize=14)
plt.xlabel('Meses', fontsize=12)
plt.ylabel('Ventas (unidades)', fontsize=12)

plt.legend()
plt.tight_layout()
plt.show()
```

```
import pandas as pd

# Cargar el CSV con separador ;
df = pd.read_csv('CSV Laboratorio 2 AD-B.csv', sep=';')

# Convertir columnas numéricas correctamente
df['Precio_unitario'] = df['Precio_unitario'].astype(str).str.replace(',', '.')
df['Precio_unitario'] = pd.to_numeric(df['Precio_unitario'], errors='coerce')
df['Cantidad_vendida'] = pd.to_numeric(df['Cantidad_vendida'], errors='coerce')

# Crear columna 'Ingresos Totales'
df['Ingresos Totales'] = df['Precio_unitario'] * df['Cantidad_vendida']

# Filtrar productos con más de 40 unidades vendidas
productos_mas_40 = df[df['Cantidad_vendida'] > 40]
print("\nProductos con más de 40 unidades vendidas:")
print(productos_mas_40)

# Filtrar los 10 precios más altos
precios_top10 = df.nlargest(10, 'Precio_unitario')
print("\nTop 10 precios unitarios más altos:")
print(precios_top10[['Producto', 'Precio_unitario']])



```

```
import pandas as pd
df = pd.read_csv('CSV Laboratorio 3 AD-B.csv')

# Limpiar encabezados
df.columns = df.columns.str.strip()

# Convertir todo a string para limpiar espacios
df = df.astype(str).apply(lambda col: col.str.strip())

# Reemplazar valores vacíos y comunes no reconocidos como nulos por pd.NA
valores_nulos = ["", " ", "nan", "NaN", "none", "None", "null", "NULL", "nan ", " NA", "NA"]
df.replace(valores_nulos, pd.NA, inplace=True)

# Convertir columnas numéricas y fechas a sus tipos reales para asegurar buen manejo
df["Value"] = pd.to_numeric(df["Value"], errors='coerce')
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')

# Estandarizar Category
df["Category"] = df["Category"].str.lower().str.replace("-", " ").str.strip()
df["Category"] = df["Category"].replace({
    "category a": "category a",
    "category b": "category b",
    "cat c": "category c",
    "cat c ": "category c",
    "cat-c": "category c"
})

# Eliminar filas con valores nulos
df = df.dropna()

# Eliminar duplicados
df = df.drop_duplicates()

# Validar limpieza final
print("Nulos por columna:")
print(df.isnull().sum())
print("Duplicados:", df.duplicated().sum())
```

```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear DataFrame
df = pd.DataFrame({
    "Horas de Estudio": [5, 10, 15, 20, 25, 30, 35, 40, 45, 50],
    "Calificación": [60, 65, 70, 75, 80, 85, 88, 90, 92, 95]
})

# Gráfico con línea de regresión
plt.figure(figsize=(8, 5))
sns.regplot(x="Horas de Estudio", y="Calificación", data=df, color="blue", ci=None)
plt.title("Relación entre Horas de Estudio y Rendimiento Académico")
plt.grid(True)

# Mostrar
plt.tight_layout()
plt.show()
```
